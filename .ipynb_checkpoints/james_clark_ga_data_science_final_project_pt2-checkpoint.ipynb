Real estate value can be extremely subjective. Almost all single family and rental properties can be summed up based on some standard objective features - square footage, number of bedrooms and bathrooms, lot size, and so on. But, there are other features that may be attractive to potential purchasers that are not as easily quantifiable.

The goal of this analysis is to use the language used to describe the property listing to determine other feature of a property listing. Additionally, a value could be associated with these objective features.

Analysis will be done using a set of Trulia property listing data found on kaggle.com (https://www.kaggle.com/promptcloud/real-estate-data-from-trulia).

```python
import pandas as pd
import stringcase
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import SnowballStemmer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import StandardScaler
from fold_to_ascii import fold
import os
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from mpl_toolkits.mplot3d import Axes3D
from scipy import stats
import numpy as np
from sklearn.model_selection import KFold
from sklearn.linear_model import LinearRegression
from sklearn import metrics

%matplotlib inline
%load_ext memory_profiler
```

```python
def pluck_number(text):
    number_only = re.sub("[^\d\.]+", "", str(text)) if pd.notnull(text) else 0
    return 0 if number_only == "" else float(number_only)
    
def drop_unused_columns(df: pd.DataFrame):    
    columns_to_keep = {'description', 'price', 'sqr_ft', 'longitude', 'latitude', 'lot_size', 'beds', 'bath'}
    df.drop(columns = df.columns.difference(columns_to_keep), inplace = True)
    
def numerize_columns(df: pd.DataFrame):    
    df.sqr_ft.fillna(0, inplace = True)
    df.lot_size.fillna(df.sqr_ft, inplace = True)
    df.beds.fillna(0, inplace = True)
    df.bath.fillna(0, inplace = True)
    for column in ['sqr_ft', 'price', 'lot_size']:
        df[column] = df[column].apply(pluck_number)
    

def remove_rows_with_empty_descriptions(df: pd.DataFrame):    
    df.dropna(subset = ['description'], inplace = True)    
    
def clean_text(text: str):
    punctuations="?:!.,;-@/*&$()_="
    sentence_words = nltk.word_tokenize(text)
    stemmer = SnowballStemmer('english')    
    custom_stop_words = ['home', 'room', 'latitude', 'longitude', 'price']
    
    words = [word.lower() for word in sentence_words if word.lower() not in stopwords.words('english') + custom_stop_words]    
    words = [word for word in words if word not in punctuations]    
    words = [fold(word) for word in words]        
    words = [stemmer.stem(word) for word in words]    
    words = [word for word in words if not re.search('(.)\1{2,}', word)] #Any word with a character repeating 3 or more times
    words = [word for word in words if not re.search('\d+', word)]    
    words = [word for word in words if not re.search('_', word)]
    words = [word for word in words if not re.search('^[^a,e,i,o,u]+$', word)]
    
    return ' '.join(words)
    
def clean_data(df: pd.DataFrame):    
    df.columns = [stringcase.snakecase(column.lower()) for column in df.columns]        
    remove_rows_with_empty_descriptions(df)    
    drop_unused_columns(df)            
    numerize_columns(df)        
    df.description = df.description.apply(clean_text)
    

def tokenize_descriptions(df: pd.DataFrame):
    vect = TfidfVectorizer(max_features = 1000)    
    text_df = pd.DataFrame(vect.fit_transform(df.description).toarray(), columns=vect.get_feature_names())    
    return text_df
    
def tokenize_descriptions_and_save(df: pd.DataFrame, force_refresh = False) -> pd.DataFrame:
    filename = 'data/cache/tokenized_descriptions.csv'
    if not force_refresh and os.path.isfile(filename):
        return pd.read_csv(filename)

    text_df = tokenize_descriptions(df)
    text_df.to_csv(filename, index = False)
    return text_df
            
def load_and_clean_train_test_data(force_refresh = False) -> pd.DataFrame:
    input_filename = 'data/home/sdf/marketing_sample_for_trulia_com-real_estate__20190901_20191031__30k_data.csv'
    cache_filename = 'data/cache/cleaned_marketing_sample_for_trulia_com.csv'
    if not force_refresh and os.path.isfile(cache_filename):
        return pd.read_csv(cache_filename)
    
    df = pd.read_csv(input_filename)
    clean_data(df)
    df.to_csv(cache_filename, index = False)    
    return df   

def reduce_dimensionality(df: pd.DataFrame, n_components: int = 2):        
    pca = PCA(n_components = n_components)
    principal_components = pca.fit_transform(df)    
    principal_df = pd.DataFrame(data = principal_components, columns = ['pca' + str(i) for i in range(1, n_components + 1)])
    return principal_df
```

## Clean the data, and tokenize the listing descriptions

#### Take the top 1000 terms - there are almost 30000 unique terms if we took all into account 

```python
#Set force_refresh to True to reload it all from scratch.
force_refresh = False
df = load_and_clean_train_test_data(force_refresh)
text_df = tokenize_descriptions_and_save(df, force_refresh)
price_df = pd.DataFrame({'index': df.index, 'price': df.price})
```

### Let's see what terms are most relevant in our listings...

```python
sorted_text_series = text_df.mean().sort_values(ascending = False)
sorted_text_df = pd.DataFrame({"terms": sorted_text_series.index, "relevance": sorted_text_series.values})

sns.catplot(data = sorted_text_df.head(15), y = 'terms', x = 'relevance', kind = 'bar')
plt.title('Most Relevant Terms')

sns.catplot(data = sorted_text_df.tail(15), y = 'terms', x = 'relevance', kind = 'bar')
plt.title('Least Relevant Terms')

plt.show()
```

### Let's figure out which features and terms have the largest impact on price

```python
meta_df = df.drop(columns = ['description', 'latitude', 'longitude', 'price'])

#Prepend "feature_" to our square footage, beds, baths, and lot size so we don't have collisions with description terms
meta_df.columns = [('feature_' + column) for column in meta_df.columns]
meta_df = pd.concat([meta_df, text_df], axis = 1)
```

```python
scaled_meta_df = pd.DataFrame(StandardScaler().fit_transform(meta_df), columns = meta_df.columns)
scaled_meta_df.head()
```

```python
correlation = scaled_meta_df.corrwith(price_df.price, method = 'kendall')

correlation_df = pd.DataFrame({'correlation': correlation.values, 'terms': correlation.index})
sorted_correlation_df = correlation_df.sort_values(by = 'correlation', ascending = False)

sns.catplot(data = sorted_correlation_df.head(20), y = 'terms', x = 'correlation', kind = 'bar')
plt.title('Terms With Greatest Correlation To Price')

sns.catplot(data = sorted_correlation_df.tail(20), y = 'terms', x = 'correlation', kind = 'bar')
plt.title('Terms With Least Correlation To Price')

sns.catplot(data = sorted_correlation_df[correlation_df.terms == 'feature_lot_size'], x = 'correlation', y = 'terms', kind = 'bar')
plt.title('Correlation of Lot Size to Price')

plt.show()
```

```python
#####
```

### Takeaways from these correlations
* As expected, square footage, bedrooms, and bathrooms have the strongest correlation with price.
* Unexpectedly, lot size has a negative correlation with price
* Terms related to additional features - such as a "suite", kitchen amenities ("island", "chef", "wine"), and "custom"/"designer" craftsmanship have some of the most positive correlations to price.
* Listings marketed as "investment opportunities", "foreclosures", or "auction" have some of the worse correlations to price.

### Building our predictive model based on popular features, and description terms

```python
decomposed_df = reduce_dimensionality(meta_df)
```

```python
for x in decomposed_df.columns:
    for zscore in [3, 1, 0.01]:    
        no_pca_outliers_decomposed_df = decomposed_df[(np.abs(stats.zscore(decomposed_df[['pca1', 'pca2']])) < zscore).all(axis = 1)]
        sns.scatterplot(data = no_pca_outliers_decomposed_df, x = x, y = price_df.price)
        plt.title(x + ' - Plot with PCA component Z-Score < ' + str(zscore))
        plt.show()
        
        
zscore = 1
no_pca_outliers_decomposed_df = decomposed_df[(np.abs(stats.zscore(decomposed_df[['pca1', 'pca2']])) < zscore).all(axis = 1)]
sns.scatterplot(data = no_pca_outliers_decomposed_df, x = 'pca1', y = 'pca2')
plt.title('PCA Terms Plot with PCA component Z-Score < ' + str(zscore))
plt.show()
```

```python
X = decomposed_df[['pca1', 'pca2']]
y = price_df.price
```

```python
best_regression = None
best_error = None
kfold = KFold(n_splits = 10, shuffle = True)
reg = None

for train_idx, test_idx in kfold.split(X):    
    train_X, test_X = X.iloc[train_idx], X.iloc[test_idx]
    train_y, test_y = y.iloc[train_idx], y.iloc[test_idx]    
    reg = LinearRegression().fit(train_X, train_y)
    predictions = reg.predict(test_X)    
    accuracy = (predictions - test_y).mean()
    error = metrics.mean_squared_error(test_y, predictions)
    if best_error == None:
        best_error = error
        best_regression = reg
    elif error < best_error:
        best_error = error
        best_regression = reg
        
```
